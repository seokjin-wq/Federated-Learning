```python
import pandas as pd

# 1. 원본 데이터 불러오기
raw_df = pd.read_csv("/content/DieCasting_Quality_Raw_Data.csv")
raw_df.columns = raw_df.columns.str.strip()

# 2. 목표 행 수 (비율 37:33:16:14)
target_counts = {'A': 3300, 'D': 460, 'B': 530, 'C': 3245}

# 3. 각 그룹의 컬럼 구조 정의
columns_map = {
    "A": ['id','Shot','Velocity_1','Velocity_2','Velocity_3','High_Velocity',
          'Cylinder_Pressure','Rapid_Rise_Time','Biscuit_Thickness','Clamping_Force','Cycle_Time',
          'Pressure_Rise_Time','Casting_Pressure','Spray_Time','Spray_1_Time','Spray_2_Time',
          'Short_Shot_1','Bubble_1','Exfoliation_1','Blow_Hole_1','Deformation_1'],
    "D": ['id','Shot','Velocity_1','Velocity_2','Velocity_3','High_Velocity',
          'Cylinder_Pressure','Rapid_Rise_Time','Biscuit_Thickness','Clamping_Force','Cycle_Time',
          'Pressure_Rise_Time','Casting_Pressure','Spray_Time','Spray_1_Time','Spray_2_Time',
          'Short_Shot_1','Exfoliation_1','Blow_Hole_1','Deformation_1'],
    "B": ['id','Shot','Velocity_1','Velocity_2','Velocity_3','High_Velocity',
          'Cylinder_Pressure','Rapid_Rise_Time','Biscuit_Thickness','Clamping_Force','Cycle_Time',
          'Pressure_Rise_Time','Casting_Pressure','Spray_Time','Spray_1_Time','Spray_2_Time',
          'Short_Shot_1','Blow_Hole_1','Exfoliation_1'],
    "C": ['id','Shot','Velocity_1','Velocity_2','Velocity_3','High_Velocity',
          'Cylinder_Pressure','Rapid_Rise_Time','Biscuit_Thickness','Clamping_Force','Cycle_Time',
          'Pressure_Rise_Time','Casting_Pressure','Spray_Time','Spray_1_Time','Spray_2_Time',
          'Short_Shot_1','Deformation_1','Exfoliation_1']
}

# 4. 데이터 무작위 섞기
shuffled_df = raw_df.sample(frac=1, random_state=42).reset_index(drop=True)

# 5. 비율대로 분할 및 컬럼 맞추기
splits_final = {}
start = 0
for key, count in target_counts.items():
    subset = shuffled_df.iloc[start:start+count].copy()
    subset = subset.reindex(columns=columns_map[key])  # 지정된 컬럼 순서대로
    splits_final[key] = subset
    start += count

# 6. 저장
for key, df in splits_final.items():
    df.to_excel(f"{key}_ratio_final.xlsx", index=False)
```
